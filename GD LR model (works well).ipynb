```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
class GD_Linear(object):
    def __init__(self,learning_rate = 0.005, n_iterations = 20):
        self.learning_rate = learning_rate
        self.n_iterations = n_iterations
    def fit(self,X,y):
        theta = np.random.randn(X.shape[1],1)
        error_cont = []
        for i in range(1,self.n_iterations+1):
            y_pred = np.dot(X, theta)
            mae = np.sum(abs(y_pred-y))/len(y)
            grads = (2/len(y)) * np.dot(X.T,np.dot(X,theta)-y)
            theta -= self.learning_rate * grads
            error_cont.append(mae)
            print(f'epoch_{i}::mae:{mae}')
        return theta,error_cont
```

```python
X1 = np.random.randn(100)
X2 = np.random.randn(100)
y = np.random.randn(100) + X1*3
```

```python
GD_Linear(learning_rate=0.01, n_iterations= 500).fit(X.reshape(100,2),y.reshape(-1,1))
```
