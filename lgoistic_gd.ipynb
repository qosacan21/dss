```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
```

```python
X, y= load_iris().data, load_iris().target.reshape(-1,1)
```

```python
X.shape, y.shape
```

```python
class SG_logistic(object):
    def __init__(self, eta = 0.005, iterations = 1000):
        self.eta = eta
        self.iterations = iterations
        
    def fit(self,X,y):
        theta = np.random.randn(X.shape[1],1)
        i = 0
        self.m = len(y)
        keep_story = []
        while(i != self.iterations):
            a_pred = np.dot(X, theta)
            y_pred = self.sigmoid(a_pred)
            y_pred = np.where(y_pred>0.5,1,0)
            cost = self.acc(y,y_pred)
            keep_story.append(cost)
            grads = (2/self.m) * np.dot(X.T,np.dot(X,theta)-y)
            theta -= self.eta * grads
            print(f'epoch{i}::{cost}::{theta}')
            i += 1
        return theta,keep_story
        
    
    def sigmoid(self,X):
        return 1 / (1+np.exp(X))
    
    def acc(self,y_true,y_pred):
        return np.sum(y_true == np.where(y_pred>0.5,1,0))/self.m
```

```python
theta, history = SG_logistic(eta = 0.000005, iterations=10000).fit(X[0:100,:],y[0:100,:])
```

```python
pd.DataFrame(history).plot(figsize = (16,5))
plt.ylim(0,1)
```
